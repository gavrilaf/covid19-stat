```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
regions = pd.read_csv('it-covid-regions-megred.csv')
```

```python
# calculate some additional data
# death amount per 100.000
# cases per 100.000
# tests performed in percent
# mortality rate (percents)
```

```python
regions['death_per_100'] = regions['death'] / (regions['population']/100000)
regions['cases_per_100'] = regions['total_cases'] / (regions['population']/100000)
regions['tests_perc'] = (regions['tests_perfomed'] / regions['population']) * 100
regions['mortality'] = (regions['death'] / regions['total_cases']) * 100
```

```python
# Sort regions by density
# Build graphics for the cases per 100.000 & death per 100.000
```

```python
regions.sort_values(by='density', ascending=False, inplace=True)
```

```python
regions.plot.bar(x='region_name', y='density', figsize=(10,5))
```

```python
regions.plot.bar(x='region_name', y='cases_per_100', figsize=(10,5))
```

```python
regions.plot.bar(x='region_name', y='death_per_100', figsize=(10,5))
```

```python
# Sort regions by population
# Build graphics for the cases per 100.000 & tests performed (in percents)
```

```python
regions.sort_values(by='population', ascending=False, inplace=True)
```

```python
regions.plot.bar(x='region_name', y='population', figsize=(10,5))
```

```python
regions.plot.bar(x='region_name', y='cases_per_100', figsize=(10,5))
```

```python
regions.plot.bar(x='region_name', y='tests_perc', figsize=(10,5))
```

```python
# Sort regions by a percentage of the population older than 65
# Build graphics for the mortality (death/cases) & death per 100.000
```

```python
regions.sort_values(by='65_', ascending=False, inplace=True)
```

```python
regions.plot.bar(x='region_name', y='65_', figsize=(10,5))
```

```python
regions.plot.bar(x='region_name', y='mortality', figsize=(10,5))
```

```python
regions.plot.bar(x='region_name', y='death_per_100', figsize=(10,5))
```
