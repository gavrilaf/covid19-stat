```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
regions = pd.read_csv('it-covid-regions-megred.csv')
```

```python
regions['dead_per_100'] = regions['dead'] / (regions['population']/100000)
```

```python
regions['tests_perc'] = (regions['tests_perfomed'] / regions['population']) * 100
```

```python
regions['mortality'] = (regions['dead'] / regions['total_cases']) * 100
```

```python
regions.sort_values(by='density', ascending=False, inplace=True)
```

```python
regions.plot.bar(x='region_name', y='density', figsize=(10,5))
```

```python
regions.plot.bar(x='region_name', y='dead_per_100', figsize=(10,5))
```

```python
regions.plot.bar(x='region_name', y='mortality', figsize=(10,5))
```

```python
regions.plot.bar(x='region_name', y='tests_perc', figsize=(10,5))
```
